```python
#do this for all fold
def create_df_for_each_fold(path, fold):
    path = path + fold + "/"
    print("path : ", path)
    file_names = os.listdir(path)
    #print(file_names)
    # Create Dictionary for File Name and Text
    file_name_and_text = {}
    for file in file_names:
        with open(path + file, "r") as target_file:
            file_name_and_text[file] = target_file.read()
    file_data = (pd.DataFrame.from_dict(file_name_and_text, orient='index')
                .reset_index().rename(index = str, columns = {'index': 'file_name', 0: 'text'}))
    return file_data            
```

```python
import os.path
from os import path

#df_list = []
result_list = []
outer_tag_folder = ["negative_polarity", "positive_polarity"]
inner_tag_folder = ["truthful_from_TripAdvisor", "deceptive_from_MTurk"]
base_path = './op_spam_training_data/'
count = 0
for outer in outer_tag_folder:
    for inner in inner_tag_folder:
        df_list = []
        path1 = base_path + outer + "/" + inner + "/"
        if path.exists(path1):
            #print("Path exist :", path1)
            pass
        else:
            path1 = base_path + outer + "/" + "truthful_from_Web" + "/"
            #print("path doesn't exist :", path1)
            

        
        for i in range(1,5):
            fold = "fold" + str(i)
            df = create_df_for_each_fold(path1, fold)
            df_list.append(df)
        result = pd.concat(df_list)
        result["category"] = str(count)  
        count = count + 1

        result_list.append(result)   

messages = pd.concat(result_list)
messages = messages.reset_index()
#print(messages)
```

```python
# Remove stop words and "!,."
# do stemming
# do tokenization [count2vector]
# Apply tfidf
import re
import string
def text_clean_1(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w','',text)
    return text
    
cleaned1 = lambda x: text_clean_1(x)    
```

```python
df2['cleaned_text'] = pd.DataFrame(df2.text.apply(cleaned1))
```

```python
#Apply a second round of cleaning
def text_clean_2(text):
    text = re.sub('[''""_]','',text)
    text = re.sub('\n','',text)
    return text
cleaned2 = lambda x: text_clean_2(x)  
```

```python
df2['cleaned_new'] = pd.DataFrame(df2.cleaned_text.apply(cleaned2))
df2.head(10)
```

```python
# Remove stop words
from nltk.corpus import stopwords  
from nltk.tokenize import word_tokenize 

def remove_stopwords(example_sent):
    stop_words = set(stopwords.words('english')) 
    word_tokens = word_tokenize(example_sent) 
    filtered_sentence = [w for w in word_tokens if not w in stop_words]  

cleaned3 = lambda x: remove_stopwords(x) 
df2['cleaned_new_stop'] = pd.DataFrame(df2.cleaned_new.apply(cleaned3))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
tvec = TfidfVectorizer()
clf2 = MultinomialNB()
from sklearn.pipeline import Pipeline
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score

print("Accuracy : ", accuracy_score(predictions, Y_test))
print("Precision : ", precision_score(predictions, Y_test,average = "weighted"))
print("Recall : ", recall_score(predictions, Y_test, average="weighted"))
```

```python
    priors = {'trust': {'truthful': 0, 'deceptive': 0},
                       'sent': {'positive': 0, 'negative': 0}}
    class_feat_counts = {'trust': {'truthful': dict(), 'deceptive': dict()},
                                  'sent': {'positive': dict(), 'negative': dict()}}


    # called by class_feat_count()
    def extract_features(doc_file, class_trust, class_sent):
        r_fd = open(doc_file, 'r')
        raw_text = r_fd.readlines()[0]
        word_list = raw_text.split(' ')
        for word in word_list:
            word1 = word.lower().strip()
            # Remove digits and punctuation
            word_list2 = word1.split(',')
            for word_i in word_list2:
                word_ii = word_i

                # Edge case: check for inter-word commas
                if word_ii == '':
                    continue

                # Update priors
                priors['trust'][class_trust] += 1
                priors['sent'][class_sent] += 1

                # Add trust feature
                if word_ii not in class_feat_counts['trust'][class_trust]:
                    # Create feature count
                    class_feat_counts['trust'][class_trust][word_ii] = 1
                else:
                    # Update feature count
                    class_feat_counts['trust'][class_trust][word_ii] += 1

                # Add sentiment feature
                if word_ii not in class_feat_counts['sent'][class_sent]:
                    # Create feature count
                    class_feat_counts['sent'][class_sent][word_ii] = 1
                else:
                    # Update feature count
                    class_feat_counts['sent'][class_sent][word_ii] += 1

                  # Compute prior_probs
        for c in priors.keys():
            print("c : ", c)
            c_token_count = sum(priors[c].values())
            print("c_token_count :", c_token_count)
            labels = list(priors[c].keys())
            print("labels :", labels)
            for l in labels:
                c_l_prob = priors[c][l] / c_token_count
                prior_probs[c][l] = c_l_prob          
```

```python
doc_file = 'op_spam_training_data/positive_polarity/truthful_from_TripAdvisor/fold2/t_talbott_9.txt'
class_trust = 'truthful'
class_sent = 'positive'
extract_features(doc_file, class_trust, class_sent)
```
